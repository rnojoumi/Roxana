```python
# load in the file
in_file= open("cocoli.txt")
#read in the first line
#the header we dont want in dict
in_file.readline()

#set up empty dict
sizes_94={}
sizes_97={}
sizes_98={}

for line in in_file:
    items = line.split()
    tree_id = items[0]
    size_in_94= float(items[4])
    size_in_97 = float(items[5])
    size_in_98= float(items[6])
    
    sizes_94[tree_id] = size_in_94
    sizes_97[tree_id] = size_in_97
    sizes_98[tree_id] = size_in_98
    
```

```python
import matplotlib.pyplot as plt

#plot some histograms
size94 = list(sizes_94.values())
plt.hist(size94, log=2)
plt.show()
```

```python
#graph both logged and unlogged
f, (ax1, ax2) = plt.subplots(1,2,sharex= True)
ax1.hist(size94, log=False)
ax1.set_title("unlogged")
ax2.hist(size94, log=True)
ax2.set_title("logged")
plt.show()
```
